```python
from terraingraph import create_graph
import math
from sanity_check import heuristic
from meshgraph import MeshGraph

n = 500

tif_path = "trentino.tif"
osm_path = "trentino_alto_adige.pbf"
graph = create_graph(tif_path=tif_path, osm_pbf_path=osm_path, resolution=n)
```

```python
i = 0
for node in graph.nodes:
    if graph.nodes[node]["is_water"] is False:
        i +=1

print(i == n**2)
```

```python
water_nodes = sum(1 for node in graph.nodes if graph.nodes[node].get("is_water", False))
land_nodes = sum(1 for node in graph.nodes if not graph.nodes[node].get("is_water", False))

print(f"Nodi totali: {len(graph.nodes)}")
print(f"Nodi acqua: {water_nodes}")
print(f"Nodi terra: {land_nodes}")
```
